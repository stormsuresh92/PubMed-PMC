```python
#!pip install requests-html
from requests_html import HTMLSession
import nest_asyncio

nest_asyncio.apply()
s = HTMLSession()

header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
        'Content-Type':'text/html',
        'Retry-After': '3600'
}
```

```python
 url = ('https://www.google.com/search?q=site%3A.ai#ip=5')
```

```python
res = s.get(url, headers=header)
print(res.status_code)
cont = res.html.find('div.N54PNb.BToiNc.cvP2Ce')
data = []
for content in cont:
  ur = content.find('a', first=True).attrs['href']
  ti = content.find('span.VuuXrf', first=True).text
  des = content.find('div.VwiC3b.yXK7lf.lyLwlc.yDYNvb.W8l4ac.lEBKkf', first=True).text
  dic = {
      'title' : ti,
      'link' : ur,
      'description' : des
  }
  data.append(dic)

```

```python
#saving to csv file
import pandas as pd
df = pd.DataFrame(data)
df.to_csv('exdata2.csv', index=False)
```
