```python
#!pip install bibtexparser
import bibtexparser

#open and read bib file
with open('test.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)

#looping all entries
data = []
for entry in bib_database.entries:
  title = entry['title']
  journal = entry['journal']
  volume = entry['volume']
  pages = entry['pages']
  year = entry['year']
  issn = entry['issn']
  doi = entry['doi']
  url = entry['url']
  author = entry['author']

 #converting data into dictionary
  dic = {
      'title':title,
      'journal':journal,
      'volume':volume,
      'pages':pages,
      'year':year,
      'issn':issn,
      'doi':doi,
      'url':url,
      'author':author
  }
  data.append(dic)
```

```python
#save to csv file format
import pandas as pd

df = pd.DataFrame(data)
df.to_csv('parsed.csv', index=False)
```
