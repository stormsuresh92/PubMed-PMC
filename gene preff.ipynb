```python
#!pip install requests-html
from requests_html import HTMLSession
#import nest_asyncio

#nest_asyncio.apply()
s = HTMLSession()

header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
        'Content-Type':'text/html',
        'Retry-After': '3600'
}
```

```python
urls = open('urls.txt', 'r')
url =urls.readlines()
```

```python
for uu in url:
  url_s = uu.strip()
  #print(url_s)
  res = s.get(url_s, headers=header)
  #print(res.status_code)
  cont = res.html.find('div.rprt-section.gene-general-protein-info')
  data = []
  for i in cont:
    try:
      name = i.find('dd', first=True).text
    except:
      name = ''

    dic = {
        'name':name,
        'link':url_s
    }
  data.append(dic)
print(data)
```
